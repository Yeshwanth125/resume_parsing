```python
import tika
from tika import parser
import re

def parseAnyFile():
        tika.initVM()
        parsed = parser.from_file("NM-Accountant-Resume-Sample.docx")
        print(parsed["content"])
        regular_expression = re.compile(r'\(([^)]*)\)', re.IGNORECASE)
        code = re.search(regular_expression,parsed["content"])
        print("code:")
        print(code)
        regular_expression2 = re.compile(r'\d{3}-\d{4}', re.IGNORECASE)
        phoneno = re.search(regular_expression2,parsed["content"])
        print('phone:')
        print(phoneno)
        regular_expression1=re.compile(r"\d*[0-9]\s?[a-zA-Z]*\s?(.*)",re.IGNORECASE)
        address = re.search(regular_expression1,parsed["content"])
        print('address:')
        print(address)
        
```

```python
parseAnyFile()
```
